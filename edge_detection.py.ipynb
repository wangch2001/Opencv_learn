```python
import cv2
import matplotlib.pyplot as plt
```

```python
# Load the image
image=cv2.imread('fig2.png',cv2.IMREAD_GRAYSCALE)

# cv2.IMREAD_COLOR: 彩色图像
# cv2.IMREAD_GRAYSCALE: 灰度图像
```

```python
# Display the original image
plt.figure(figsize=(6,6))
plt.imshow(image,cmap='gray')
plt.title('Original Image')
plt.axis('off')
plt.show()
```

```python
# Sobel
sobel_x=cv2.Sobel(image,cv2.CV_64F,1,0,ksize=3)
sobel_y=cv2.Sobel(image,cv2.CV_64F,0,1,ksize=3)
sobel=cv2.magnitude(sobel_x,sobel_y)
```

```python
# Scharr
scharr_x=cv2.Scharr(image,cv2.CV_64F,1,0)
scharr_y=cv2.Scharr(image,cv2.CV_64F,0,1)
scharr=cv2.magnitude(scharr_x,scharr_y)
```

```python
# Laplacian
laplacian=cv2.Laplacian(image,cv2.CV_64F)
```

```python
# Canny
canny=cv2.Canny(image,50,150)
```

```python
# Prewitt
prewitt_kx=cv2.getDerivKernels(1,0,3,normalize=True)
prewitt_ky=cv2.getDerivKernels(0,1,3,normalize=True)
prewitt_x=cv2.sepFilter2D(image,cv2.CV_64F,prewitt_kx[0],prewitt_kx[1])
prewitt_y=cv2.sepFilter2D(image,cv2.CV_64F,prewitt_ky[0],prewitt_ky[1])
prewitt=cv2.magnitude(prewitt_x,prewitt_y)
```

```python
# Display the results
fig,axs=plt.subplots(2,3,figsize=(15,10))
axs=axs.ravel()

methods=[sobel,scharr,laplacian,canny,prewitt]
titles=['Sobel','Scharr','Laplacian','Canny','Prewitt']

for i in range(5):
    axs[i].imshow(methods[i],cmap='gray')
    axs[i].set_title(titles[i])
    axs[i].axis('off')
    
axs[5].axis('off')
plt.tight_layout()
plt.show()
```
